```python
import pandas as pd
import os
```

```python
flist = os.listdir('./ptthomework')
flist
```

```python
data = []
for textfile in flist:
    if textfile == '.ipynb_checkpoints': 
        continue    #遇到 textfile = .ipynb_checkpoints,就跳過不做
    with open('./ptthomework/%s' %(textfile),'r',encoding='utf-8') as f:
        data_str = f.read().split('---split---')[1]
        data_row = data_str.split('\n')[1:-1]
        data.append(data_row)
        #print(data_row)
        #print(data_str)
    #print('-------------------------------------------')
data
```

```python
columns = ['推','噓','作者','標題','時間']
df =pd.DataFrame(data=data,columns=columns)
df
```

```python
#製造過濾器
def tmpfilter(s):    #將字串(s)放入名為tmpfilter的過濾器
    result = s.split(':')[1]  #使用:分隔,取第一個
    return result   #return回去
```

```python
tmpfilter('推:4')  #測試,過濾後只剩數字
```

```python
df_copied = df.copy()   #做備份
```

```python
df_copied['推']
```

```python
df_copied['推'].apply(tmpfilter)
```

```python
df_copied['推'] = df_copied['推'].apply(tmpfilter)   #取出要的欄位,放入過濾器
```

```python
df_copied
```

```python
df_copied = df.copy() 
df_copied
```

```python
columns
```

```python
#使用迴圈將欄位放入過濾器
for c in columns:
    df_copied[c] = df_copied[c].apply(tmpfilter) 
df_copied
```

```python
#重寫另一個
df_copied2 = df.copy()
df_copied2
```

```python
#不使用過濾器,改使用lambda
for c in columns:
    df_copied2[c] = df_copied2[c].apply(lambda s: s.split(':')[1]) 
df_copied2
```

```python
#儲存成csv檔
df_copied2.to_csv('./ptt.csv',index=False)
```
