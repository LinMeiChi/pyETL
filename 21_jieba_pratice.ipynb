```python
import jieba
import os
jieba.load_userdict("./mydict.txt")
```

```python
#取標題
flist = os.listdir('./PyETL/ptthomework')
flist 
```

```python
#取內容
sb= ""
for textfile in flist:
    if textfile == '.ipynb_checkpoints':
        continue
    with open("./PyETL/ptthomework/%s" %(textfile), "r",encoding="utf-8") as f:
        sb += f.read().split("---split---")[0]
sb
```

```python
#利用 jieba(預設模式)分割
s_cut = jieba.cut(sb)
```

```python
for w in s_cut:
    print(w)
```

```python
s_cut = jieba.cut(sb)
wc = {}   #自訂空字典

for w in s_cut:    #從段詞內抓
    if len(w) < 2:  #長度小於2
        continue    #略過
    if not w in wc:  #如果不在裡面,為1
        wc[w] = 1
    else:    #否則,+1
        wc[w] += 1
wc
```

```python
wc_list = [(w,wc[w])for w in wc]
wc_list
```

```python
wc_list = []
for w in wc:
    tmpdata = tuple ([w,wc[w]])
    wc_list.append(tmpdata)
wc_list
```

```python
#做排序(預設由小排到大),採用reverse做翻轉
wc_list.sort(key=lambda t: t[1],reverse=True)
wc_list
```
